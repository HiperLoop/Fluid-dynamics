3.a Lists  

A list is a group of items (strings, numbers, booleans) grouped together under one name. 
They are extremely flexible, but this also makes computations with them slow. 
Lists are not suitable to store, for example, field data (like the temperature at each grid point). Nor can they be used for vector operations.

More info: https://earth-env-data-science.github.io/lectures/core_python/python_fundamentals.html 

```python
list1 = ('hello', 3, True) #you can combine very different things here... separate by comma 

numbers1=(1, 2, 3)
numbers2=(4, 5, 6)

numbers1+numbers2  #this is concatenation, not addition!! 
```

```python
# indexing is about grabbing certain entries (located at a certain index value) from a list... 
# ... or from other items such as arrays! 

# python indexing might be confusing at first, so always be careful to doublecheck - do a little test in an extra cell, if needed! 

list1=('a','b','c','d','e')

N=len(list1) #the length of the list (nr of entries)
print('length of list = ',N )

print('   individual values')
print(list1[1]) #the first index is 0; so the index 1 takes the second entry! 
print(list1[-2])  #use negative signs to count backwards. -1 is the last, -2 the second last

print('   ranges')
print(list1[1:3]) #use : for ranges of values. 1:3 means start with index 1 (the second entry)
                  # and go to "just before" index 3 (i.e. up to index 2, the third entry)
print(list1[:3]) # from zero up to and including index 2   - equivalent to [0:3] 
print(list1[-3:]) # from the third-last to the last - equivalent to [-3:-1]

print('   skipping values')
print(list1[0:-1:2]) #skip every second item. we go to "just before" the last entry (-1)
print(list1[0:5:2]) #now we go to "just before" 5, i.e. we do include the last entry (index 4)
```

4. Repetition and Conditional Commands 

4.a Repeting operations: for-loops

Often you want to do the same thing over and over again (e.g. for every time step, for every location etc). This can be done with a for-loop. 


```python
#an example of a for-loop 

for i in range(0,5):   # i is the index and will go from 0 to 4 here. 
    j=i*2
    for k in range(0,2):
        print(k)
    print(i,j)           # indent everything within the loop by 4 spaces or 1 tab 
    
print("I am done with the loop!")     #stop indenting when exiting the loop 
```

Note: loops take relatively long to execute, especially nested loops. If possible, use vector operations. 

(I have seen students writing a code that took 15 minutes to run, where a more efficient, but certainly not state-of-the-art implementation took 15 seconds!)

```python
# two ways of adding two vectors: loop vs vector operation

a=np.array([1, 2, 3, 4, 5])
b=np.linspace(0.1,0.5,5)   # vector with linear interpolation between 0.1 and 0.5 with 5 entries

# the inefficient way - using a loop over all elements
c=np.zeros(5,)       # vector with all entries 0
for i in range(0,len(b)): 
    c[i]=a[i]+b[i]
    
print('this is vector c', c)    


# the efficient way - using vector operation (here, addition)
d=a+b
    
print('this is vector d', d) 
```

4.b conditional statements

```python
# a few if statements 


x = 0
if x > 0:            # the first possible case
    print('x>0') 
elif x < 0:          # "else if"-statement: if the first codition does not hold, but the second...
    print('x<0')
else:                # "else" : if none of the previous conditions holds.... 
    print ('Zero!')
    
print("done with the first conditional")


# an example of combining conditions 
a=1; b=3; c=True 

if a<5 and b<5 or c:
    print("   at least one of the following holds")
    print("   1.   a and b are both smaller than 5")
    print("   2.   condition c is True")

```

5. Functions 

Functions take input and return output. 
The nice thing is that you can define them somewhere in your notebook and reuse them elsewhere, rather than repeating similar orders all the time. 

```python
# example function

# first, define the function.

def velocity_local (x,y,t,a,om):
    #this function gives the velocity in position (x,y) and time t
    #a is the amplitude, om the frequency
    
    pi=np.pi   #define some auxiliary variable that we will not return 
    f=2*pi*om
    
    u=a*   x*np.cos(f*t)  #velocity in x-direction
    v=a*(-y)*np.cos(f*t)  # and y-direction
    
    return u,v  # decide which variables the function will give to the main code 

# everything below "def" is inside the function, so long as we indent the lines. 

# next, use the function 

u_local, v_local = velocity_local(1,2,0,1,3)
print(u_local,v_local)
    
```

6 Plotting 

First, we make a line plot of two time series of velocities.

Second, we visualise fields

More info: https://earth-env-data-science.github.io/lectures/basic_scipy/more_matplotlib.html 

```python
# for plotting, we import matplotlib: 
from matplotlib import pyplot as plt 

# now, use the velocity function to generate a time series ar position (1,1) and plot that
x0=1
y0=1
a=1
om=1

tmin=0; #begin time
tmax=3; #end time 
nt=61; #nr of time points (nr of intervalls = nr time points -1)

tt=np.linspace(tmin,tmax,nt)   #time points

ut=np.zeros(nt,)  #these will contain the time series data 
vt=np.zeros(nt,)

for it in range(0,nt): 
    t=tt[it]
    uu,vv = velocity_local(x0,y0,t,a,om)
    ut[it]=uu
    vt[it]=vv

    
# now, do the plotting    
fig, ax = plt.subplots()  #make fig with 1 item (called ax)
ax.plot(tt,ut,label='u')    
ax.plot(tt,vt,label='v')   #plot v and give it a label (for the legend)

ax.set_xlabel('time')    #axis labels  and figure title
ax.set_ylabel('u,v')
ax.set_title('velocities')
leg=ax.legend(loc='upper right') #activate legend and locate it
```

```python
# now let's generate velocity field data for a given time step and plot that 

xmin=-3    #fix your domain: smallest and highest coordinate values
xmax=3
ymin=-3
ymax=3
nx=301    #nr of points in each direction
ny=301

xlin = np.linspace(xmin, xmax, nx)    #generate vectors of x and y values
ylin = np.linspace(ymin, ymax, ny)
xx, yy = np.meshgrid(xlin, ylin)      #combine to 2 matrices holding x and y values for each grid point
    
(uu,vv) = velocity_local(xx,yy,t,a,om)  #use the velocity function to generate velocity values



fig = plt.figure(figsize=(12,6))       #define a figure 
axes = fig.subplots(nrows=1, ncols=2)  # axes are the subplots (in this case 2)
ax0,ax1=axes                           # name the subplots ax0 and ax1 

name0=ax0.pcolormesh(xx,yy,uu, shading='auto')                           # plot the component uu in the first subplot
ax0.set_title('u')
fig.colorbar(name0,ax=ax0)

name1=ax1.pcolormesh(xx,yy,vv, shading='auto')                           # plot the component vv in the second subplot
ax1.set_title('v')
fig.colorbar(name1,ax=ax1)
```

```python
#for velocity fields, it can be nicer to make a quiver plot. make sure not to plot too many data points for readability 

nskip=15     
uur=uu[0:nx:nskip,0:ny:nskip]   #only keep every nskip-th value 
vvr=vv[0:nx:nskip,0:ny:nskip]
xxr=xx[0:nx:nskip,0:ny:nskip]
yyr=yy[0:nx:nskip,0:ny:nskip]

fig, ax = plt.subplots()  #make fig with 1 item (called ax)
ax.quiver(xxr,yyr, uur, vvr)                          # visualiseer u en v door pijltjes
```
